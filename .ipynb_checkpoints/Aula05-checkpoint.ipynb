```python
import matplotlib.pyplot as plt
x = [1,2,3,4] #Passo 1
y = [10,20,25,30]
fig = plt.figure() #Passo 2
ax = fig.add_subplot(1,1,1) #Passo 3 adiciona um axes (grafico, imagem, etc)
ax.plot(x, y, color='lightblue', linewidth=3) #Passo 4
ax.scatter([2,4,6], [5,15,25], color='darkgreen', marker='^')
ax.set_xlim(1, 6.5)
plt.savefig('foo.png') #Passo 5
plt.show() #Passo 6
```

### 3. Rotinas de plotagem
### Dados 1D

```python
fig, ax = plt.subplots()
# Plota pontos conectados por linhas e marcadores
lines = ax.plot(x, y)
# Plota pontos desconectados, em escala ou coloridos
ax.scatter(x, y)
# Plota ret칙ngulos verticais (comprimento constante)
axes[0,0].bar([1,2,3], [3,4,5])
# Plota ret칙ngulos horizontais (altura constante)
axes[1,0].barh([0.5,1,2.5], [0,1,2])
# Plota uma linha horizontal atrav칠s dos eixos
axes[1,1].axhline(0.45)
# Plota uma linha vertical atrav칠s dos eixos
axes[0,1].axvline(0.65)
# Plota pol칤gonos preenchidos
ax.fill(x, y, color='blue')
# Preenche entre valores-y e 0
ax.fill_between(x, y, color='yellow')
```

```python
#importando as classes da forma mais usual
import matplotlib.pyplot as plt
import numpy as np
#definindo vari치veis usadas nos exemplos
pop = {"S칚o Paulo": 12000000, "Rio": 7000000, "Salvador": 4000000, "Recife":
3500000, "Porto Alegre": 2500000}
cidades = [i for i in pop.keys()]
populacao = [j for j in pop.values()]
popPos = np.arange(len(cidades))
azul = "blue"
verde = "green"
preto b= "black"
```

### Desenhando um gr치fico de barras horizontal

```python
def plotaBarraH1():
    plt.barh(popPos, populacao, align='center', color=azul)
    plt.yticks(popPos, cidades)
    plt.xlabel('Popula칞칚o')
    plt.title('Popula칞칚o de cidades brasileiras')
```

```python
plotaBarraH1()
```

### Desenhando um gr치fico de duas barras

```python
def plotaBarDupla1():
    grupos = 5
    media_chico = (9.1, 5.4, 4.0, 7.5, 7.0)
    media_joao = (8.6, 6.3, 5.5, 3.5, 5.6)
    fig, ax = plt.subplots()
    indice = np.arange(grupos)
    bar_larg = 0.4
    transp = 0.7
    plt.bar(indice, media_chico, bar_larg, alpha=transp, color=azul,
label='Chico')
    plt.bar(indice + bar_larg, media_joao, bar_larg, alpha=transp, color=verde,
label='Jo칚o')

    plt.xlabel('Mat칠ria')
    plt.ylabel('Notas')
    plt.title('Notas por pessoa')
    plt.xticks(indice + bar_larg, ('Matem치tica', 'Portugu칡s', 'Biologia', 'F칤sica',
'Qu칤mica'))
    plt.legend() #for칞a os labels aparecem
    plt.tight_layout()
    plt.show() #necessario para exibir a figura
```

```python
plotaBarDupla1()
```

### Desenhando um gr치fico de pizza

```python
def plota_pizza_1():
    labels = ['Corinthias', 'Palmeiras', 'Santos', 'S칚o Paulo']
    titulos = [27, 22, 22, 22]
    cores = ['lightblue', 'green', 'white', 'red']
    explode = (0.1, 0, 0, 0) # somente explode primeiro peda칞o
    total = sum(titulos)
    plt.pie(titulos, explode=explode, labels=labels, colors=cores, autopct=lambda
    p: '{:.0f}'.format(p * total / 100), shadow=True, startangle=90)
    # Determina que as propor칞칫es sejam iguais ('equal') de modo a desenhar o
    #c칤rculo
    plt.axis('equal')
    plt.show()
```

```python
plota_pizza_1()
```

### Desenhando um gr치fico de dispers칚o (Scatter Plot)

```python
def plotaDispersao1():
    fig = plt.figure()
    fig2=plt.figure()
    ax1 = fig.add_subplot(121)
    ax2 = fig2.add_subplot(121)
    #gera dados aleat칩rios
    N=1000
    x = np.random.randn(N)
    y = np.random.randn(N)
    ax1.scatter(x,y,color=azul,s=5,edgecolor='none')
    #ax1.plot(x,y, "bo")
    #deixa os eixos quadrados
    ax1.set_aspect(1./ax1.get_data_ratio())
    ax2.plot(x,y, "b.")
    ax2.set_aspect(1./ax2.get_data_ratio())
    plt.show()
```

```python
plotaDispersao1()
```
